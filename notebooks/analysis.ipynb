# Stock ETL & ML Prediction Analysis
This notebook provides a quick walkthrough of extracting AAPL stock data, transforming it, loading it to SQLite, running SQL queries, and training the Random Forest Classifier.

```python
import sys
import os
# Adjust path to import custom modules
sys.path.append(os.path.abspath('..'))

import pandas as pd
import sqlite3
```

## 1. Run ETL Pipeline
Extract data from yfinance, transform it, and load it into SQLite database.

```python
from etl.extract import extract_stock_data
from etl.load import run_etl

# Step 1: Download raw CSV data
extract_stock_data()

# Step 2: Clean, transform and load into database
db_path = run_etl()
print(f"Database created at: {db_path}")
```

## 2. Load and Query Data via SQLite

```python
conn = sqlite3.connect('../database/stocks.db')
df = pd.read_sql_query("SELECT * FROM stocks ORDER BY Date ASC", conn)
conn.close()

print(f"Data loaded. Total Rows: {len(df)}")
df.head()
```

```python
from model.predict import train_and_predict

results = train_and_predict(db_path='../database/stocks.db')
print(f"Test Set Accuracy: {results['accuracy']:.2%}")
print(f"Tomorrow's Prediction: {results['prediction']} (Confidence: {results['confidence']:.2f}%)")
```
